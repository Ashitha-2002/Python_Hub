```python
class employee():
    def __init__(self,empid, name, designation, salary):
        self.empid = empid
        self.name = name
        self.designation = designation
        self.salary = salary

    def getEmpDetails(self):
        print("\n----------------------------------------\n")
        print("Emp Id :: ", self.empid)
        print("Employee Name :: ",self.name)
        print("DEsignation :: ", self.designation)
        print("Salary :: ", self.salary)
        print("Gross Salary :: ", self.getGrossSal())
        print("Net Salary :: ", self.getNetSal())

    def getGrossSal(self):
        DA = 5 / 100 * self.salary
        HRA = 10 / 100 * self.salary
        gross = self.salary + DA + HRA
        return gross

    def getNetSal(self):
        LIC = 3 / 100 * self.salary
        pf = 8 / 100 * self.salary
        deduction = LIC + pf
        net = self.getGrossSal() - deduction
        return net

empid = []
name = []
designation = []
salary = []
for i in range(5):
    empid.append(int(input("Enter empid : ")))
    name.append(input("Enter employee name : "))
    designation.append(input("Enter Designation : "))
    salary.append(float(input("Enter Salary : ")))
    emp = employee(empid[i], name[i], designation[i], salary[i])
for i in range(5):
    emp.getEmpDetails()
```

```python
from multipledispatch import dispatch
class example:
    @dispatch(int, int)
    def add(self, a, b):
        print("int int")
        return a + b

    @dispatch(float, float)
    def add(self, a, b):
        print("float float")
        return a + b

    @dispatch(int, int, int)
    def add(self, a, b, c):
        print("int int int")
        return a + b + c

    @dispatch(float, float, float)
    def add(self, a, b, c):
        print("float float float")
        return a + b + c

    @dispatch(int, float)
    def add(self, a, b):
        print("int float")
        return a + b

    @dispatch(str, str)
    def add(self, a, b):
        print("str str")
        return a + b

obj = example()

print(obj.add(5,20,3))
print(obj.add(6,1))
print(obj.add(2.1, 0.5))
print(obj.add(5, 7, 3))
print(obj.add(10, 0.7))
print(obj.add("Good", " Afternoon"))
```

```python
class Teacher:
    def __init__(self, name, age, subject):
        self.name = name
        self.age = age
        self.subject = subject

    def display(self):
        print("Name: ",self.name,"\nAge:", self.age, "\nSubject:" ,self.subject)


class Professor(Teacher):
    def __init__(self, name, age, subject, research_area):
        super().__init__(name, age, subject)
        self.research_area = research_area

    def display(self):
        super().display()
        print("Research Area: ",self.research_area)


class Ast_Professor(Teacher):
    def __init__(self, name, age, subject, years_of_experience):
        super().__init__(name, age, subject)
        self.years_of_experience = years_of_experience

    def display(self):
        super().display()
        print("Years of Experience:", self.years_of_experience)


class As_Professor(Teacher):
    def __init__(self, name, age, subject, publications):
        super().__init__(name, age, subject)
        self.publications = publications

    def display(self):
        super().display()
        print("Publications:", self.publications)

professor = Professor("Dr. John", 50, "Computer Science", "Data Science")
ast_professor = Ast_Professor("David", 32, "Physics", 5)
as_professor = As_Professor("Dr. Williams", 45, "Chemistry", ["Journal of Chemistry", "Chemical Reviews"])

print("Professor Information:")
professor.display()
print("\nAssociate Professor Information:")
as_professor.display()
print("\nAssistant Professor Information:")
ast_professor.display()
```
